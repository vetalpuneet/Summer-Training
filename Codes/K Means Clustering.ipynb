```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv('Customers.csv')
```

```python
dataset
```

```python
x = dataset.iloc[:, [3, 4]].values
x
```

```python
# using elbow method to find optimal number of clusters
#wcss is within cluster sum of squares and help to create elbow method
# initialization of centroids and their is optimized value find out, is basically cost fun and which ever is lower cost fun that perticular cntroid will be considered
#inertia to give wcss value of each clustor as loop goes
# wcss = summ of euc dist btw the scattered point
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans=KMeans(n_clusters=i, init='k-means++',random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clustors')
plt.ylabel('WCSS')
plt.show()
```

```python
# Fitting kmeans to the dataset

kmeans=KMeans(n_clusters=5,init='k-means++',random_state=0)
y_kmeans=kmeans.fit_predict(x)
y_kmeans
```

```python
# visualize the clustors
#s = size pf pixel
plt.scatter(x[y_kmeans==0,0],x[y_kmeans==0,1],s=100,c='red',label='Cluster1')
plt.scatter(x[y_kmeans==1,0],x[y_kmeans==1,1],s=100,c='blue',label='Cluster2')
plt.scatter(x[y_kmeans==2,0],x[y_kmeans==2,1],s=100,c='green',label='Cluster3')
plt.scatter(x[y_kmeans==3,0],x[y_kmeans==3,1],s=100,c='cyan',label='Cluster4')
plt.scatter(x[y_kmeans==4,0],x[y_kmeans==4,1],s=100,c='magenta',label='Cluster5')

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='yellow',label='Centroids')

plt.title('Clusters of customers')
plt.xlabel('Annual Income(K$)')
plt.ylabel('Spending Score(1-100)')
plt.legend()
plt.show()
```
